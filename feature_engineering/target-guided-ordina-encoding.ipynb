## Target Guided Ordinal Encoding

Target Guided Ordinal Encoding (TGOE) is a technique for encoding high-cardinality categorical features by replacing each category with the **mean of the target variable** for that category. Unlike One-Hot Encoding, it adds only one new column regardless of how many unique categories exist, making it well-suited for features with many distinct values.

**How it works:**
1. Compute the mean target value for each category
2. Map each category to its corresponding mean
3. The resulting column carries both the encoding and a signal about the relationship between the category and the target

**Important:** Mean values must be computed on the **training set only**. Computing them on the full dataset leaks target information into the features and will inflate model performance.

```python
import pandas as pd
df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'New York', 'Chicago', 'New York', 'Houston', 'Phoenix'],
    'price': [250, 150, 300, 250, 320, 280, 200]
})
```

```python
df
```

```python
mean_price = df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded'] = df['city'].map(mean_price)
```

```python
df
```

```python
df[['price', 'city_encoded']].drop_duplicates()
```

## Conclusion

Target Guided Ordinal Encoding is a compact, signal-rich alternative to One-Hot Encoding for high-cardinality categorical features.

| | One-Hot Encoding | Ordinal Encoding | Target Guided Ordinal Encoding |
|---|---|---|---|
| New columns added | One per category | 1 | 1 |
| Encodes natural order | No | Yes (if specified) | No |
| Uses target signal | No | No | Yes |
| Risk of data leakage | No | No | Yes (if not careful) |

**When to use TGOE:**
- The categorical feature has many unique values (high cardinality)
- A meaningful relationship exists between the category and the target
- You want to avoid the dimensionality explosion of OHE

**Always fit the encoding on the training set only**, then apply it to the validation and test sets to prevent data leakage.
