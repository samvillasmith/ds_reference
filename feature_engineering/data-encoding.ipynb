## Data Encoding

Encoding converts categorical variables into a numerical format that machine learning algorithms can process. This notebook covers:

1. One-Hot Encoding (OHE)
2. Label and Ordinal Encoding
3. Target Guided Ordinal Encoding

---

### One-Hot Encoding (OHE)

One-Hot Encoding creates a new binary column for each unique category. A row gets a  in the column matching its category and  in all others.



Best suited for **nominal** (unordered) categories. Avoid for high-cardinality features, as each unique value adds a new column.

---

### Sparse vs Dense Representation

By default,  returns a **sparse matrix** — only the non-zero values are stored, which is memory-efficient when most entries are zero.

Calling  converts it to a **dense array**, where all values (including zeros) are stored explicitly. Dense arrays are easier to work with but use more memory.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, OrdinalEncoder
```

```python
# Create a dataframe
data = {'Color': ['Red', 'Blue', 'Green', 'Red', 'Blue']}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# Create an instance of OneHotEncoder
ohe = OneHotEncoder()
```

```python
# Perform fit and transform the data
encoded_data = ohe.fit_transform(df[['Color']]).toarray()
```

```python
encoded_data
```

```python
import pandas as pd
encoder_df = pd.DataFrame(encoded_data, columns=ohe.get_feature_names_out())
```

```python
encoder_df
```

```python
ohe.transform([['Blue']]).toarray()
```

```python
pd.concat([df, encoder_df], axis=1)
```

## Label Encoding

Label encoding assigns each unique category a unique integer. The assignment is alphabetical by default.



**Note:** Label encoding implies an ordinal relationship between categories (0 < 1 < 2), which can mislead models into treating categories as ordered. Use it only for the **target variable** or genuinely binary features — not for nominal input features like color.

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
```

```python
lbl_encoder.fit_transform(df['Color'])
```

```python
lbl_encoder.transform(['Green'])
```

```python
lbl_encoder.transform(['Blue'])
```

## Ordinal Encoding
Ordinal encoding is similar to label encoding, but it takes into account the order of the categories. For example, if we have a variable "Size" with categories "Small", "Medium", and "Large", we could assign "Small" = 0, "Medium" = 1, and "Large" = 2, reflecting the natural order of the sizes.

```python
df = pd.DataFrame({'Size': ['Small', 'Medium', 'Large', 'Medium', 'Small', 'Large', 'Medium', 'Small']})
```

```python
df
```

```python
# Create an instance of OrdinalEncoder
ord_encoder = OrdinalEncoder(categories=[['Small', 'Medium', 'Large']])
```

```python
ord_encoder.fit_transform(df[['Size']])
```

```python
ord_encoder.transform([['Medium']])
```

## Conclusion

Choosing the right encoding strategy depends on whether your feature has a natural order and on the model you are using.

| Method | Use When | Watch Out For |
|---|---|---|
| **One-Hot Encoding** | Nominal data, no natural order | High-cardinality features produce too many columns |
| **Label Encoding** | Target variable or binary features | Implies false ordinal relationships on nominal data |
| **Ordinal Encoding** | Ordered categories (Small < Medium < Large) | Order must be explicitly specified — default is alphabetical |

**Key takeaways:**
- Use OHE for nominal input features with low-to-moderate cardinality
- Use Ordinal Encoding when a meaningful order exists between categories
- Avoid Label Encoding on nominal input features — the implied numeric ordering can introduce bias in linear and distance-based models
